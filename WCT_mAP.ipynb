# Weighted Cross Testing, WCT

1) Row-wise, from each method we get 4 test sets - light, moderate, heavy and severe. Each test set contains 500 images. 

 Based on the 4 test sets, we have S = [S1, S2, ..., S15], mAP = [ mAP1, mAP2, ..., mAP15], weight = [w1， w2, ..., w15]. Then we can calculate the WCT_row as below：

```python
# S = [S1, S2, ..., S15]

weight = [1/32, 1/32, 1/32, 1/32, 2/32, 2/32, 2/32, 2/32, 2/32, 2/32, 3/32, 3/32, 3/32, 3/32, 4/32] # 共15个权重值

# mAP = [mAP1, mAP2, ..., mAP15] # 共15个mAP值
mAP =  [
    0.242, 0.227, 0.22, 0.197, 0.232, 0.228, 0.222, 0.211,
    0.207, 0.206, 0.226, 0.215, 0.213, 0.211, 0.216
]


WCT = 0
for i in range(len(mAP)):
    WCT += mAP[i] * weight[i]

print("WCT_row =", WCT)

```

2) Column-wise, for each weather severity we have 2 test sets, for example light_method_1, light_method_2. Each test set contains 500 images. 

 Based on the 2 test sets, we have S = [S1, S2, S3], mAP = [ mAP1, mAP2, mAP3], weight = [w1 w2, w3]. Then we can calculate the WCT_column as below：

```python
# S = [S1, S2, S3]

weight = [1/4, 1/4, 2/4] # 共3个权重值

# mAP = [mAP1, mAP2, mAP3] # 共3个mAP值
mAP = [0.072, 0.012, 0.056]



WCT = 0
for i in range(len(mAP)):
    WCT += mAP[i] * weight[i]

print("WCT_column=", WCT)
```
